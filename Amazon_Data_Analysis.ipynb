```python
df = df.dropna()
df
```

```python
df.isnull().sum()
```

```python
print(df.dtypes)
```

```python
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False)
df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False)
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '', regex=False).str.replace(',', '', regex=False)
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')
df['rating'] = df['rating'].replace(['|', '-', '', ' '], np.nan)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '', regex=False)
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce').astype('Int64')  
print(df.dtypes)
```

```python
df
```

```python
df =  df.drop(columns=['review_id','review_title', 'review_content','img_link','product_link'])
df
```

```python
df_unique = df.drop_duplicates(subset='product_name')
df_rating = df_unique.dropna(subset=['rating'])
top_10 = df_rating.sort_values(by=['rating', 'rating_count'], ascending=[False, False]).head(10).reset_index(drop=True)
top_10 
```

```python
top_10['short_name'] = top_10['product_name'].apply(lambda x: x[:30] + '...' if len(x) > 30 else x)
plt.figure(figsize=(14, 7))
sns.barplot(data=top_10, y='short_name', x='rating')
plt.title('Top 10 Products by Rating', fontsize=15)
plt.xlabel('Rating')
plt.ylabel('Product Name')
plt.tight_layout()
plt.show()
```
