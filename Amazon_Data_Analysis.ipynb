```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv(r"C:\Users\PC\Desktop\DataSet\amazon.csv")
df
```

```python
df = df.dropna()
df
```

```python
df.isnull().sum()
```

```python
print(df.dtypes)
```

```python
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False)
df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False)
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '', regex=False).str.replace(',', '', regex=False)
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')
df['rating'] = df['rating'].replace(['|', '-', '', ' '], np.nan)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '', regex=False)
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce').astype('Int64')  
print(df.dtypes)
```

```python
df
```

```python
df =  df.drop(columns=['review_id','review_title', 'review_content','img_link','product_link'])
df
```

```python
df_unique = df.drop_duplicates(subset='product_name')
df_rating = df_unique.dropna(subset=['rating'])
top_10 = df_rating.sort_values(by=['rating', 'rating_count'], ascending=[False, False]).head(10).reset_index(drop=True)
top_10 
```

```python
top_10['short_name'] = top_10['product_name'].apply(lambda x: x[:30] + '...' if len(x) > 30 else x)
plt.figure(figsize=(14, 7))
sns.barplot(data=top_10, y='short_name', x='rating')
plt.title('Top 10 Products by Rating', fontsize=15)
plt.xlabel('Rating')
plt.ylabel('Product Name')
plt.tight_layout()
plt.show()
```

```python
df_unique = df.drop_duplicates(subset='product_name')
most_discounted = df_unique.sort_values(by='discount_percentage', ascending=False).head(10)
most_discounted.reset_index(drop=True, inplace=True)
most_discounted['short_name'] = most_discounted['product_name'].str.slice(0, 40)
most_discounted
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data= most_discounted, x='short_name', y='discount_percentage')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Product')
plt.ylabel('Discount Percentage')
plt.title('Top 10 Products with Highest Discount', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
cheapest_actual = df[['product_name', 'actual_price']].dropna().drop_duplicates()
cheapest_actual_10 = cheapest_actual.nsmallest(10, 'actual_price')
cheapest_actual_10['short_name'] = cheapest_actual_10['product_name'].str.slice(0, 30) + '...'
cheapest_actual_10 = cheapest_actual_10.reset_index(drop=True)
cheapest_actual_10
```

```python
plt.figure(figsize=(10, 6))
plt.hlines(y=cheapest_actual_10['short_name'], xmin=0, xmax=cheapest_actual_10['actual_price'], color='skyblue', linewidth=2)
plt.plot(cheapest_actual_10['actual_price'], cheapest_actual_10['short_name'], "o", color='steelblue')
plt.xlabel('Actual Price (₹)')
plt.title('Top 10 Cheapest Products by Actual Price', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Prepare Data
most_reviewed = df[['product_name', 'rating_count']].dropna().drop_duplicates()
most_reviewed = most_reviewed.sort_values(by='rating_count', ascending=False).head(10)
most_reviewed['short_name'] = most_reviewed['product_name'].str.slice(0, 30) + '...'
most_reviewed = most_reviewed.reset_index(drop=True)
most_reviewed
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=most_reviewed, y='short_name', x='rating_count')
plt.title('Top 10 Most Reviewed Products', fontsize=15)
plt.xlabel('Number of Ratings')
plt.ylabel('Product')
plt.tight_layout()
plt.show()
```

```python
category_counts = df['category'].value_counts().head(10).reset_index(drop=True)
category_counts
```

```python
plt.figure(figsize=(8, 8))
colors = sns.color_palette('pastel')[0:10]
plt.pie(category_counts, labels=category_counts.index, colors=colors, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})
plt.title('Top 10 Product Categories Distribution', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")
fig = plt.figure(figsize=(22, 20))  # Dashboard size
ax1 = plt.subplot2grid((3, 2), (0, 0))
sns.barplot(data=top_10, y='short_name', x='rating', ax=ax1)
ax1.set_title('Top 10 Products by Rating', fontsize=13)
ax1.set_xlabel('Rating')
ax1.set_ylabel('Product Name')
ax2 = plt.subplot2grid((3, 2), (0, 1))
sns.barplot(data=most_discounted, x='short_name', y='discount_percentage', ax=ax2)
ax2.set_title('Top 10 Products with Highest Discount', fontsize=13)
ax2.set_xlabel('Product')
ax2.set_ylabel('Discount %')
ax2.tick_params(axis='x', rotation=45)
ax3 = plt.subplot2grid((3, 2), (1, 0))
ax3.hlines(y=cheapest_actual_10['short_name'], xmin=0, xmax=cheapest_actual_10['actual_price'], color='skyblue', linewidth=2)
ax3.plot(cheapest_actual_10['actual_price'], cheapest_actual_10['short_name'], "o", color='steelblue')
ax3.set_title('Top 10 Cheapest Products by Actual Price', fontsize=13)
ax3.set_xlabel('Actual Price (₹)')
ax3.grid(axis='x', linestyle='--', alpha=0.7)
ax4 = plt.subplot2grid((3, 2), (1, 1))
sns.barplot(data=most_reviewed, y='short_name', x='rating_count', ax=ax4)
ax4.set_title('Top 10 Most Reviewed Products', fontsize=13)
ax4.set_xlabel('Number of Ratings')
ax4.set_ylabel('Product')
ax5 = plt.subplot2grid((3, 2), (2, 0), colspan=2)
colors = sns.color_palette('pastel')[0:10]
ax5.pie(category_counts, labels=category_counts.index, colors=colors, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})
ax5.set_title('Top 10 Product Categories Distribution', fontsize=14)
plt.tight_layout()
plt.show()
```
